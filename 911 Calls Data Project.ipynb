## Creating new features

** creating a new column called "Reason" that contains the reason of the call.** 

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df,palette='viridis')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

** creating 3 new columns called Hour, Month, and Day of Week.**

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
df['Day of Week'].head()
```

** Notice how the Day of Week is an integer 0-6. Changing it to the actual string names to the day of the week.**

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

** creating a countplot of the Day of Week column with the hue based off of the Reason column. **

```python
sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

** doing the same for the month**

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

** We notice it is missing some months! 9,10, and 11 are not there. **

** Now creating a gropuby object called byMonth **

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

** Creating a new column called 'Date' that contains the date from the timeStamp column.**

```python
df['Date']=df['timeStamp'].apply(lambda t: t.date())
```

** groupby this Date column with the count() aggregate and creating a plot of counts of 911 calls.**

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

** Now recreating this plot with 3 separate plots, each plot representing a Reason for the 911 call**

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```


** restructuring the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. **

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

** Now creating a HeatMap using this new DataFrame. **

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

** Now creating a clustermap using this DataFrame. **

```python
sns.clustermap(dayHour,cmap='viridis')
```

** repeating these same plots and operations, for a DataFrame that shows the Month as the column. **

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
sns.clustermap(dayMonth,cmap='viridis')
```
